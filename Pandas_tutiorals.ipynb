```python
### every region top 10 item 
df['Percent Of Region Sales'] =df["sales"]/df.groupby('region')['sales'].transform('sum')
df.head()
```

```python
df["rank"]=df.groupby('region')['sales'].rank(ascending=False)
```

```python
# Filtering Rows Where the Group's Average Sale Price is Less Than 20,000
df=df.groupby('region').filter(lambda x:x["sales"].mean()<20000)
```

```python
df.groupby(['gender','region']).sum()
```

```python
df.groupby(['gender','region'])["sales"].rank(ascending=False)
```

```python
def group_range(x):
  return x.max()-x.min()
```

```python
range=df.groupby(["region", "gender"])['sales'].apply(group_range)
print(range)
```
